```python
#ВІЗУАЛІЗАЦІЯ ГР.1
#Графік мов програмування
import gdown
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('2017_jun_final.csv')

plt.figure(figsize=(12, 8))
sns.countplot(x='Язык.программирования', data=data, order=data['Язык.программирования'].value_counts().index, palette='viridis')
plt.title('Розподіл вакансій за мовами програмування')
plt.xlabel('Мова програмування')
plt.ylabel('Кількість вакансій')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#ВІЗУАЛІЗАЦІЯ ГР.2
#Графік зарплатні для м.п. Python 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('2017_jun_final.csv')

python_data = data[data['Язык.программирования'] == 'Python']

plt.figure(figsize=(12, 8))
sns.histplot(x='Зарплата.в.месяц', data=python_data, bins=30, kde=True, color='pink')
plt.title('Розподіл зарплат для вакансій з Python')
plt.xlabel('Зарплатня')
plt.ylabel('Кількість вакансій')
plt.show()
```

```python
##ВІЗУАЛІЗАЦІЯ ГР.3
#Графік співвідношення Junior, Middle та Senior Software Engineer
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Читання даних з файлу CSV та заміна українських символів в іменах стовпців
data = pd.read_csv('2017_jun_final.csv', encoding='utf-8')
data.columns = data.columns.str.replace('і', 'i')

positions_data = data[(data['Должность'].isin(['Junior Software Engineer', 'Middle Software Engineer', 'Senior Software Engineer'])) & 
                      (data['Язык.программирования'].isin(['Python', 'Java']))]

plt.figure(figsize=(10, 6))
sns.countplot(x='Должность', data=positions_data, hue='Язык.программирования', palette='viridis')
plt.title('Розподіл вакансій за позиціями та мовами програмування')
plt.xlabel('Позиція')
plt.ylabel('Кількість вакансій')
plt.legend(title='Мова програмування')
plt.show()
```

```python
#14.Створіть новий DataFrame, де для згрупованих даних за стовпчиком "Посада", виконайте агрегацію даних за допомогою методу agg і знайдіть мінімальне та максимальне значення у стовпчику "Зарплата.в.місяць"
import pandas as pd
import gdown

data = pd.read_csv('2017_jun_final.csv')

grouped_data = data.groupby('Должность')

agg_result = grouped_data.agg({
    'Зарплата.в.месяц': ['min', 'max']
})

# Виведення результату
print(agg_result)
```

```python
#15.Створіть функцію fill_avg_salary, яка повертатиме середнє значення заробітної плати на місяць.
#Використовуйте її для методу apply та створіть новий стовпчик "avg"
import pandas as pd
import gdown

data = pd.read_csv('2017_jun_final.csv')

# Функція для перетворення значень у числа та обчислення середнього значення
def fill_avg_salary(row):
    numeric_values = pd.to_numeric(row, errors='coerce')
    return numeric_values.mean()

# Створення нового стовпця 'avg', який містить середнє значення всіх стовпців у кожному рядку
data['avg'] = data.apply(fill_avg_salary, axis=1)

# Виведення результату
print(data[['avg']])
```
