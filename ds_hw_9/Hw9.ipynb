```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split

# Завантаження та передобробка даних
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

# Нормалізація даних
X_train = X_train / 255.0
X_test = X_test / 255.0

# Перетворення міток у формат категоріальних змінних
y_train = tf.keras.utils.to_categorical(y_train, num_classes=10)
y_test = tf.keras.utils.to_categorical(y_test, num_classes=10)

# Побудова та налаштування моделі
model = Sequential([
    Flatten(input_shape=(28, 28)),  # Вхідний шар
    Dense(512, activation='relu'),  # Перший прихований шар з 512 нейронами та функцією активації ReLU
    Dropout(0.5),  # Dropout для захисту від перенавчання
    Dense(256, activation='relu'),  # Додатковий прихований шар з 256 нейронами та функцією активації ReLU
    Dense(128, activation='relu'),  # Третій прихований шар з 128 нейронами та функцією активації ReLU
    Dense(10, activation='softmax')  # Вихідний шар з 10 нейронами та функцією активації Softmax
])

# Компіляція
optimizer = Adam(learning_rate=0.0005)  # Зменшимо швидкість навчання
model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# Навчання
history = model.fit(X_train, y_train, epochs=40, batch_size=128, validation_split=0.2, callbacks=[EarlyStopping(patience=7)])

# Оцінка
test_loss, test_acc = model.evaluate(X_test, y_test)

print(f'Test Accuracy: {test_acc}')
```
